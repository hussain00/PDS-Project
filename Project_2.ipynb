```python
from google.colab import files

uploaded = files.upload()

file_1= list(uploaded.keys())[0]
file_2 = list(uploaded.keys())[1]
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv(file_1)
df2 = pd.read_csv(file_2)
```

```python
mean1 = df1['Marks'].mean()
mean2 = df2['Marks'].mean()
```

```python
df1['Marks'] = df1['Marks'].fillna(mean1)
df2['Marks'] = df2['Marks'].fillna(mean2)
```

```python
df = pd.merge(df1, df2[['Registration Number', 'Marks']], on='Registration Number')
```

```python
df
```

```python
df['Best One'] = df[['Marks_x','Marks_y']].max(axis=1)
df['Average'] = df[['Marks_x','Marks_y']].mean(axis=1)
```

```python
df
```

```python
df = df.drop(['Marks_x','Marks_y'], axis=1)
```

```python
from google.colab import files

uploaded = files.upload()

file_att = list(uploaded.keys())[0]
file_fin = list(uploaded.keys())[1]
```

```python
att = pd.read_csv(file_att)
tf = pd.read_csv(file_fin)
```

```python
data = df.copy()
```

```python
data = pd.merge(data, att[['Registration Number', 'Attendance Marks']], on='Registration Number')
```

```python
data
```

```python
data = pd.merge(data, tf[['Registration Number', 'Final Exam Marks']], on='Registration Number')
```

```python
data['Final Marks'] = data['Final Exam Marks']*0.7 + data['Average'] + data['Attendance Marks']
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
result_data = pd.read_csv('final result.csv')
```

```python
result_data.info()
```

```python
result_data.isnull().sum()
```

```python
result_data = result_data.iloc[:,1:]
```

```python
result_data.info()
```

```python
#Term Test Average & Total Marks

X1 = result_data.drop(columns=['Registration Number','Name','Best One','Attendance Marks','Final Exam Marks'], axis=1).values
```

```python
X1
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=X1[:,0], y=X1[:, 1])
plt.xlabel('TT Marks')
plt.ylabel('Final Marks')
plt.show()
```

```python
wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=2)
  kmeans.fit(X1)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
Y1 = kmeans.fit_predict(X1)
Y1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X1[Y1==0,0], X1[Y1==0,1], s=50, c='red', label='Cluster 1')
plt.scatter(X1[Y1==1,0], X1[Y1==1,1], s=50, c='pink', label='Cluster 2')
plt.scatter(X1[Y1==2,0], X1[Y1==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X1[Y1==3,0], X1[Y1==3,1], s=50, c='green', label='Cluster 4')
plt.scatter(X1[Y1==4,0], X1[Y1==4,1], s=50, c='orange', label='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='black', label='Centroids')
plt.title('Customer Groups')
plt.xlabel('TT')
plt.ylabel('Final Marks')
plt.show()
```
