```python
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
f_data = df[['Survived','Sex','Age','Fare']]
```

```python
print(f_data)
```

```python
f_data = f_data[f_data['Survived'] == 1]
```

```python
f_data = f_data.drop(columns=['Survived'])
```

```python
f_data
```

```python
f_data.loc[:, 'Sex'] = f_data['Sex'].apply(lambda x: 0 if x == 'male' else 1)
```

```python
f_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
colors = f_data['Sex'].map({0: 'green', 1: 'red'})

plt.scatter(f_data['Age'], f_data['Fare'], color=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot showing relation between Age and Fare')
plt.legend(handles=[
    plt.Line2D([], [], color='green', marker='o', linestyle='', label='Male'),
    plt.Line2D([], [], color='red', marker='o', linestyle='', label='Female')
])
plt.show()
```
